## Dumping to json

```python
csvfile = open('midas.csv', 'r')
jsonfile = open('myfile.json', 'w')

fieldnames = ("id","created_at","#favorites","#retweet","text","#images")
reader = csv.DictReader( csvfile, fieldnames)
for row in reader:
    json.dump(row, jsonfile)
    jsonfile.write('\n')
```
